## Load Dataset

```python
import pandas as pd

# Load the Superstore dataset
file_path = "/content/SampleSuperstore.csv"
data = pd.read_csv(file_path)
```

```python
# Display unique values in categorical columns
categorical_columns = ['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub-Category']
for column in categorical_columns:
    print(f"\nUnique values in {column}:")
    print(data[column].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize categorical columns
plt.figure(figsize=(18, 12))
for i, column in enumerate(categorical_columns, 1):
    plt.subplot(3, 3, i)
    sns.countplot(x=column, data=data, palette='viridis')
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
# Explore numerical columns
numerical_columns = ['Sales', 'Quantity', 'Discount', 'Profit']

# Summary statistics
print("\nSummary statistics of numerical columns:")
print(data[numerical_columns].describe())
```

```python
# Visualize numerical columns
plt.figure(figsize=(18, 6))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(1, 4, i)
    sns.histplot(data[column], bins=20, kde=True, color='skyblue')
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
# Check correlation between numerical columns
correlation_matrix = data[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
```

```python
# Visualize correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Sales and Profit analysis by Region
plt.figure(figsize=(16, 8))
sns.scatterplot(x='Sales', y='Profit', hue='Region', data=data, palette='viridis')
plt.title('Sales and Profit Analysis by Region')
plt.show()
```

```python
# Sales and Profit analysis by Category and Sub-Category
plt.figure(figsize=(18, 10))
sns.scatterplot(x='Sales', y='Profit', hue='Category', size='Sub-Category', sizes=(30, 300), data=data, palette='viridis')
plt.title('Sales and Profit Analysis by Category and Sub-Category')
plt.show()
```

```python
# Profit distribution by Segment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Segment', y='Profit', data=data, palette='viridis')
plt.title('Profit Distribution by Segment')
plt.show()
```

```python
# Profit distribution by Ship Mode
plt.figure(figsize=(10, 6))
sns.boxplot(x='Ship Mode', y='Profit', data=data, palette='viridis')
plt.title('Profit Distribution by Ship Mode')
plt.show()
```

```python
# Explore the relationship between Discount and Profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=data, color='purple')
plt.title('Relationship between Discount and Profit')
plt.show()
```
